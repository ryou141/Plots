# Exercise 05 : Boxplot




```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('../data/ab-test.csv')
df
```

## Отрисовка графика




```python
fig, axes = plt.subplots(1, 2, figsize=(12, 6))

titles = ["after", "before"]

for i, time in enumerate(titles):
    ax = axes[i]

    data = df[df["time"] == time]

    groups = data.groupby("group")["diff"]
    
    ax.boxplot([groups.get_group("control"), groups.get_group("test")], 
               tick_labels=["control", "test"], 
               boxprops=dict(color="green"), 
               whiskerprops=dict(color="red"), 
               capprops=dict(color="red"), 
               medianprops=dict(color="black"))
    
    ax.set_title(time)
    
fig.suptitle("Lag before labs deadlines. After and before the Newsfeed")
plt.show()
```
